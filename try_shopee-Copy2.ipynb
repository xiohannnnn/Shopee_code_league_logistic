```python
def get_city2(buyeraddress):
    if re.search("metro manila",buyeraddress,re.I):
        return 'metro manila'
    elif re.search("luzon",buyeraddress,re.I):
        return 'luzon'
    elif re.search("visayas",buyeraddress,re.I):
        return 'visayas'
    else:
        return 'mindanao'
```

```python
df['buyer_city'] = df['buyeraddress'].apply(get_city2)
df['seller_city'] = df['selleraddress'].apply(get_city2)
df.head()
```

```python
def get_SLA(buyer_city,seller_city):
    if buyer_city == 'metro manila' and seller_city =='metro manila':
        return 3
    elif buyer_city == 'metro manila' and seller_city =='luzon':
        return 5
    elif seller_city == 'metro manila' and buyer_city =='metro manila':
        return 3
    elif seller_city == 'metro manila' and buyer_city =='luzon':
        return 5
    elif seller_city == 'luzon' and buyer_city =='luzon':
        return 5
    else:
        return 7
```

```python
df['SLA'] = np.vectorize(get_SLA)(df['buyer_city'], df['seller_city'])
```

```python
def get_sunday(pickup,first_deli):
    if pickup <= datetime.date(2020,3,1) and first_deli >= datetime.date(2020,3,1):
        return 1
    elif pickup <=  datetime.date(2020,3,15) and first_deli >= datetime.date(2020,3,15):
        return 1
    elif pickup <=  datetime.date(2020,3,22) and first_deli >= datetime.date(2020,3,22):
        return 1
    elif pickup <=  datetime.date(2020,3,29) and first_deli >= datetime.date(2020,3,29):
        return 1
    elif pickup <=  datetime.date(2020,4,5) and first_deli >= datetime.date(2020,4,5):
        return 1
    else:
        return 0
```

```python
def get_ph(pickup,first_deli):
    if pickup <=  datetime.date(2020,3,8) and first_deli >= datetime.date(2020,3,8):
        return 1
    elif pickup <=  datetime.date(2020,3,25) and first_deli>=datetime.date(2020,3,25):
        return 1
    elif pickup <=  datetime.date(2020,3,30) and first_deli >= datetime.date(2020,3,30):
        return 1
    elif pickup <=  datetime.date(2020,3,31) and first_deli >= datetime.date(2020,3,31):
        return 1
    else:
        return 0
```

```python
df['public_holi'] = np.vectorize(get_ph)(df['pickup'], df['1st_deli'])
df['public_holi2'] = np.vectorize(get_ph)(df['1st_deli'], df['2nd_deli'])
#df['public_holi_total']=df['public_holi']+df['public_holi2']
df['sunday'] = np.vectorize(get_sunday)(df['pickup'], df['1st_deli'])
df['sunday2'] = np.vectorize(get_sunday)(df['1st_deli'], df['2nd_deli'])
#df['sun_total']=df['sunday']+df['sunday2']
df.head()
```
